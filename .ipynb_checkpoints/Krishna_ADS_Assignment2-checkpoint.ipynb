```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
from numpy import cov
import seaborn as sns

#read the dataset 
df_CO2emission = pd.read_excel("CO2 emissions (kt).xls")
df_CO2emission_gF_consumption = pd.read_excel("CO2 emissions from gaseous fuel consumption (kt) .xls")
df_CO2emission_lF_consumption = pd.read_excel("CO2 emissions from liquid fuel consumption (kt) .xls")
df_CO2emission_sF_consumption = pd.read_excel("CO2 emissions from solid fuel consumption (% of total).xls")
df_ForestArea = pd.read_excel("Forest area.xls")
df_Population = pd.read_excel("Population growth.xls")
```

```python
#Select different countries from CO2 Emission
slt_UK_CO2     = df_CO2emission.loc[df_CO2emission['Country Name'] == 'United Kingdom']
slt_China_CO2  = df_CO2emission.loc[df_CO2emission['Country Name'] == 'China']
slt_Spain_CO2  = df_CO2emission.loc[df_CO2emission['Country Name'] == 'Spain']
slt_Poland_CO2 = df_CO2emission.loc[df_CO2emission['Country Name'] == 'Poland']
slt_World_CO2  = df_CO2emission.loc[df_CO2emission['Country Name'] == 'World']
```

```python
#Select different countries from CO2 Emission from gaseous fuel consumption
slt_UK_CO2_gf     = df_CO2emission_gF_consumption.loc[df_CO2emission_gF_consumption['Country Name'] == 'United Kingdom']
slt_China_CO2_gf  = df_CO2emission_gF_consumption.loc[df_CO2emission_gF_consumption['Country Name'] == 'China']
slt_Spain_CO2_gf  = df_CO2emission_gF_consumption.loc[df_CO2emission_gF_consumption['Country Name'] == 'Spain']
slt_Poland_CO2_gf = df_CO2emission_gF_consumption.loc[df_CO2emission_gF_consumption['Country Name'] == 'Poland']
slt_World_CO2_gf  = df_CO2emission_gF_consumption.loc[df_CO2emission_gF_consumption['Country Name'] == 'World']
```

```python
#Select different countries from CO2 Emission from liquid fuel consumption
slt_UK_CO2_lf     = df_CO2emission_lF_consumption.loc[df_CO2emission_lF_consumption['Country Name'] == 'United Kingdom']
slt_China_CO2_lf  = df_CO2emission_lF_consumption.loc[df_CO2emission_lF_consumption['Country Name'] == 'China']
slt_Spain_CO2_lf  = df_CO2emission_lF_consumption.loc[df_CO2emission_lF_consumption['Country Name'] == 'Spain']
slt_Poland_CO2_lf = df_CO2emission_lF_consumption.loc[df_CO2emission_lF_consumption['Country Name'] == 'Poland']
slt_World_CO2_lf  = df_CO2emission_lF_consumption.loc[df_CO2emission_lF_consumption['Country Name'] == 'World']
```

```python
#Select different countries from CO2 Emission from solid fuel consumption
slt_UK_CO2_sf     = df_CO2emission_sF_consumption.loc[df_CO2emission_sF_consumption['Country Name'] == 'United Kingdom']
slt_China_CO2_sf  = df_CO2emission_sF_consumption.loc[df_CO2emission_sF_consumption['Country Name'] == 'China']
slt_Spain_CO2_sf  = df_CO2emission_sF_consumption.loc[df_CO2emission_sF_consumption['Country Name'] == 'Spain']
slt_Poland_CO2_sf = df_CO2emission_sF_consumption.loc[df_CO2emission_sF_consumption['Country Name'] == 'Poland']
slt_World_CO2_sf  = df_CO2emission_sF_consumption.loc[df_CO2emission_sF_consumption['Country Name'] == 'World']
```

```python
#Select different countries forest area
slt_UK_FA     = df_ForestArea.loc[df_ForestArea['Country Name'] == 'United Kingdom']
slt_China_FA  = df_ForestArea.loc[df_ForestArea['Country Name'] == 'China']
slt_Spain_FA  = df_ForestArea.loc[df_ForestArea['Country Name'] == 'Spain']
slt_Poland_FA = df_ForestArea.loc[df_ForestArea['Country Name'] == 'Poland']
slt_World_FA  = df_ForestArea.loc[df_ForestArea['Country Name'] == 'World']
```

```python
#Select different Countries population 
slt_UK_PL     = df_Population.loc[df_Population['Country Name'] == 'United Kingdom']
slt_China_PL  = df_Population.loc[df_Population['Country Name'] == 'China']
slt_Spain_PL  = df_Population.loc[df_Population['Country Name'] == 'Spain']
slt_Poland_PL = df_Population.loc[df_Population['Country Name'] == 'Poland']
slt_World_PL  = df_Population.loc[df_Population['Country Name'] == 'World']
```

```python
############## CO2 Emission rate of United kingdom in year2015 #####################

UK_CO2_data = pd.concat([slt_UK_CO2_gf,slt_UK_CO2_lf,slt_UK_CO2_sf],axis=0) # concantenate the three indecators of UK
UK_CO2_data.drop(UK_CO2_data.iloc[:, 61:62], inplace = True, axis = 1) # drop the column data
UK_CO2_data = UK_CO2_data.set_index('Indicator Name') #set the column name as indicator name

# Set the fontsize
font1     = {'family':'serif','color':'black','size':20}
textprops = {"fontsize":15}

# pie chart
plt.figure()
plt.pie(UK_CO2_data["2015"], labels=UK_CO2_data.index,autopct='%1.2f%%',textprops=textprops)
plt.title("CO2 Emission rate of United kingdom in year2015 ",fontsize=16, fontdict=font1, loc='center')
plt.show()

############### CO2 Emission rate of China in year2015 ##########################

China_CO2_data = pd.concat([slt_China_CO2_gf, slt_China_CO2_lf, slt_China_CO2_sf], axis=0) ## concantenate the three indecators of China
China_CO2_data.drop(China_CO2_data.iloc[:, 61:62], inplace=True, axis=1) # drop the column data
China_CO2_data = China_CO2_data.set_index('Indicator Name')  #set the column name as indicator name

# Pie chart
plt.figure()
plt.pie(China_CO2_data["2015"], labels=China_CO2_data.index, autopct='%1.2f%%',textprops=textprops)
plt.title("CO2 Emission rate of China in year2015 ", fontsize=16, fontdict=font1, loc='center')
plt.show()

################ CO2 Emission rate of Spain in year2015 #########################

Spain_CO2_data = pd.concat([slt_Spain_CO2_gf,slt_Spain_CO2_lf,slt_Spain_CO2_sf],axis=0) # concantenate the three indecators of Spain
Spain_CO2_data.drop(Spain_CO2_data.iloc[:, 61:62], inplace = True, axis = 1) # drop the column data
Spain_CO2_data = Spain_CO2_data.set_index('Indicator Name')  #set the column name as indicator name

#pie chart
plt.figure()
plt.pie(Spain_CO2_data["2015"], labels=Spain_CO2_data.index, autopct='%1.2f%%',textprops=textprops)
plt.title("CO2 Emission rate of Spain in year2015 ",fontsize=16, fontdict=font1, loc='center')
plt.show()

################ CO2 Emission rate of Poland in year 2015 ########################

Poland_CO2_data = pd.concat([slt_Poland_CO2_gf,slt_Poland_CO2_lf,slt_Poland_CO2_sf],axis=0)# concantenate the three indecators of Poland
Poland_CO2_data.drop(Poland_CO2_data.iloc[:, 61:62], inplace = True, axis = 1) # drop the column data
Poland_CO2_data = Poland_CO2_data.set_index('Indicator Name')  #set the column name as indicator name

#pie chart
plt.figure()
plt.pie(Poland_CO2_data["2015"], labels=Poland_CO2_data.index, autopct='%1.2f%%',textprops=textprops)
plt.title("CO2 Emission rate of Poland in year 2015 ",fontsize=16, fontdict=font1, loc='center')
plt.show()

################ CO2 Emission rate of World in year 2015 ###########################

World_CO2_data = pd.concat([slt_World_CO2_gf,slt_World_CO2_lf,slt_World_CO2_sf],axis=0)# concantenate the three indecators of world
World_CO2_data.drop(World_CO2_data.iloc[:, 61:62], inplace = True, axis = 1) # drop the column data
World_CO2_data = World_CO2_data.set_index('Indicator Name')  #set the column name as indicator name

#pie chart
plt.figure()
plt.pie(World_CO2_data["2015"], labels=World_CO2_data.index, autopct='%1.2f%%',textprops=textprops)
plt.title("CO2 Emission rate of World in year 2015 ",fontsize=16, fontdict=font1, loc='center')
plt.show()
```

```python
slt_UK_tr     = slt_UK_CO2.transpose() #transpose the raw with country name United Kingdom
slt_China_tr  = slt_China_CO2.transpose() #transpose the raw with country name China
slt_Spain_tr  = slt_Spain_CO2.transpose() #transpose the raw with country name Spain
slt_Poland_tr = slt_Poland_CO2.transpose() #transpose the raw with country name Poland


#Concantenate the each data with different counties in colomnwise.
Concantenate_UK_C_S_P = pd.concat([slt_UK_tr, slt_China_tr, slt_Spain_tr, slt_Poland_tr], axis=1)  

Concantenate_UK_C_S_P.drop(labels=["Country Code","Indicator Name", "Indicator Code",'2019','2020'], 
                           inplace = True, axis = 0)  # Drop the raws

# Set the header
new_header =Concantenate_UK_C_S_P.iloc[0] #grab the first row for the header
Concantenate_UK_C_S_P = Concantenate_UK_C_S_P[1:] #take the data less the header row
Concantenate_UK_C_S_P.columns = new_header

#plot the CO2 emission rate of different countries
Concantenate_UK_C_S_P.iloc[30:62].plot(kind="bar", figsize=(10, 5),stacked =True);
plt.title("CO2 Emission rate in UK, China, Spain and Poland",fontsize=16)
plt.legend( prop={"size":16})
plt.xlabel("Year",fontsize=16)
plt.ylabel("CO2 emission rate",fontsize=16)

#Concantenate_UK_C_S_P.iloc[1:40].plot.hist(alpha=0.5,figsize=(10, 10));
#Concantenate_UK_C_S_P.iloc[1:40].plot.hist(alpha=0.5, bins=50,figsize=(10, 10));
```
